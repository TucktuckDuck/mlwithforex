```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
parameters = { 'C':np.arange(1,11,0.5)}
svc = svm.SVC(class_weight='balanced')
# svc = svm.SVC()
SVM=GridSearchCV(svc, parameters)
SVM.fit(x_train,labels_train)
SVM.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
rforest_clf = RandomForestClassifier(n_jobs=-1, n_estimators=1000, class_weight="balanced").fit(x_train, labels_train)

# Evaluate the model's accuracy
print("For Random Forest model:")
print("Train set accuracy = " + str(rforest_clf.score(x_train, labels_train)))
print("Test set accuracy = " + str(rforest_clf.score(x_test, labels_test)))

# Determine features' importances
print("\nImportance of each feature:\n", rforest_clf.feature_importances_)
```
